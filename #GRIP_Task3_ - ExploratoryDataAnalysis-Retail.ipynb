### Ship Mode

```python
SS['Ship Mode'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Ship Mode',data=SS,palette='seismic')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.title('Ship Mode Countplot')
plt.show()
```

### Which Ship Mode has maximum sales and profit?

```python
SS.groupby(['Ship Mode'])[['Sales','Profit']].sum().sort_values(ascending=False,by=['Sales','Profit'])
```

### Which Region has maximum sales and profit?

```python
SS.groupby(['Region'])[['Sales','Profit']].sum().sort_values(ascending=False,by=['Sales','Profit'])
```

```python
plt.figure(figsize=(30,7))
plt.subplot(1,2,1)
sns.barplot(x=SS['Region'],y=SS.Sales,hue=SS['Ship Mode'],estimator=np.sum,ci=None,palette='rocket')
plt.title('Sales according to Ship Modes and Region')
```

```python
plt.figure(figsize=(30,7))
plt.subplot(1,2,2)
sns.barplot(x=SS['Region'],y=SS.Profit,hue=SS['Ship Mode'],estimator=np.sum,ci=None,palette='mako')
plt.title('Profit according to Ship Modes and Region')
plt.show()
```

### What are the different segments present in the data ? How much Sales and Profit is generated ? Show all these results Regionwise.

### Segment

```python
SS['Segment'].value_counts()
```

```python
plt.figure(figsize=(14,6))
sns.countplot(x='Segment',data=SS,palette='rocket_r')
plt.xlabel('Segment')
plt.ylabel('Count')
plt.title('Segment Countplot')
plt.show()
```

```python
SS.groupby(['Segment'])[['Sales','Profit']].sum().sort_values(ascending=False,by=['Sales','Profit'])
```

```python
plt.figure(figsize=(30,7))
plt.subplot(1,2,2)
sns.barplot(x=SS['Region'],y=SS.Profit,hue=SS['Segment'],estimator=np.sum,ci=None,palette='viridis_r')
plt.title('Profit according to Segment and Region')
plt.show()
```

### What is the relation between Profit generated and Discount given ?

```python
plt.figure(figsize=(15,6))
sns.lineplot(x='Discount',y='Profit',data=SS,estimator=np.mean,ci=None,color='purple')
plt.grid()
plt.title('Profit Vs Discount')
plt.show()
```

### Quantity Sales and Profit w.r.t Category

```python
SS.groupby(['Category','Sub-Category'])[['Sales','Profit','Quantity']].sum().sort_values(ascending=False,
                                                                                                 by=['Category','Profit'])
```

### Pie Chart

```python
# pie chart of Category
cat_data = SS.Category.value_counts()
cat_labels = np.unique(SS.Category)

colors = sns.color_palette('icefire')[0:4]

plt.figure(figsize=(20,20))
plt.subplot(1,2,1)
plt.pie(cat_data, labels = cat_labels, colors = colors, autopct='%.0f%%')
plt.title('Category')

# pie chart of Sub-category
subcat_data = SS['Sub-Category'].value_counts().sort_index()
subcat_labels = np.unique(SS['Sub-Category'])

colors = sns.color_palette('rainbow_r')[0:18]

plt.subplot(1,2,2)
plt.pie(subcat_data, labels = subcat_labels, colors = colors, autopct='%.0f%%')
plt.title('Sub-Category')

plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x=SS['Sub-Category'],y=SS.Profit,hue=SS['Category'],estimator=np.sum,ci=None,dodge=False)
plt.xticks(rotation=90)
plt.title('Profit according to Category and Sub-Category')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x=SS['Sub-Category'],y=SS.Sales,hue=SS['Category'],estimator=np.sum,ci=None,dodge=False)
plt.xticks(rotation=90)
plt.title('Sales according to Category and Sub-Category')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x=SS['Sub-Category'],y=SS.Quantity,hue=SS['Category'],estimator=np.sum,ci=None,dodge=False)
plt.xticks(rotation=90)
plt.title('Quantity sold by Category and Sub-Category')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x=SS['Sub-Category'],y=SS.Discount,hue=SS['Category'],estimator=np.mean,ci=None,dodge=False)
plt.xticks(rotation=90)
plt.title('Discount according to Category and Sub-Category')
plt.legend(loc='upper left')
plt.show()
```

In <b>Furniture</b> Category, <b>Chairs</b> has maximum <b>Profit</b> & <b>Sales</b>. However, <b>Furnishings</b> has maximum <b>Quantity</b> Sold and maximum <b>Discount</b> is given to <b>Tables</b>.
<br>
In <b>Office Supplies</b> Category, <b>Paper</b> has maximum <b>Profit</b>, <b>Storage</b> has maximum <b>Sales</b> and <b>Binders</b> has maximum <b>Quantity</b> Sold and <b>Discount</b> given.
<br>
In <b>Technology</b> Category, <b>Copiers</b> has maximum <b>Profit</b>, <b>Phones</b> has maximum <b>Sales</b> and <b>Quantity</b> Sold and <b>Machines</b> has maximum <b>Discount</b> given.
<br>
Overall <b>Copiers</b> has maximum <b>Profit</b>, <b>Phones</b> has maximum <b>Sales</b> and <b>Binders</b> has maximum <b>Quantity</b> Sold and <b>Discount</b> given.

### What are Profit and Sales generated according to different States ?

```python
plt.figure(figsize=(14,6))
sns.lineplot(x='State',y='Profit',data=SS,estimator=np.sum,ci=None,color='red')
plt.xticks(rotation=90)
plt.grid()
plt.title('Profit by State')
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.lineplot(x='State',y='Sales',data=SS,estimator=np.sum,ci=None,color='forestgreen')
plt.xticks(rotation=90)
plt.grid()
plt.title('Sales by State')
plt.show()
```

So from above analysis we can say that the maximum <b>Profit</b> and <b>Sales</b> is generated from <b>State - California</b> and lowest <b>Profit</b> from <b>State - Texas</b> and lowest <b>Sales</b> from <b>State - North Dakota</b>.
